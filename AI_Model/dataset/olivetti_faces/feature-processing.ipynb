# Olivetti Faces dataset

## Processing Features of Images

```python
import os
import sys
AI_MODEL_PATH = os.path.abspath(os.path.join(os.path.curdir, os.pardir, os.pardir))
sys.path.append(AI_MODEL_PATH)
from image_preprocessing import preprocess, visualize_image
import numpy as np
```

## Load the Data

```python
# https://www.kaggle.com/code/serkanpeldek/face-recognition-on-olivetti-dataset
olivetti_faces_data = np.load('olivetti_faces.npy')
olivetti_faces_data.shape
```

## Raw Image (Before Processing)

```python
visualize_image(olivetti_faces_data)
```

## Preprocessing Stage

```python
# Preprocess the data
images = []
for image in olivetti_faces_data:
    image = preprocess(
        img_array=image,
        isCurrentGrayScale=True
    )
    images.append(image)
images = np.array(images)
images.shape
```

## After Processing

```python
visualize_image(images)
```

```python
# Save the preprocessed data
if not os.path.exists('features'):
    os.mkdir('features')
np.save('features/olivetti_faces_preprocessed.npy', images)
print('Processed Data saved successfully!')
```
